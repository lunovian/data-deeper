```python
# Import necessary libraries
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

# Fetch Bitcoin price data from Yahoo Finance
btc = yf.Ticker("BTC-USD")
data = btc.history(period="max")

# Parse the data
dates = data.index
prices = data['Close']

# Apply a rolling average to smooth the data
rolling_window = 30  # 30-day rolling average
prices_smooth = prices.rolling(window=rolling_window).mean()
# plt.show()

# [BEGIN OF CUSTOMIZATION CODE]
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))
plt.plot(dates, prices_smooth, label='Bitcoin Price (USD)', color='#8B4513', linewidth=2.5)  # SaddleBrown color
plt.fill_between(dates, prices_smooth, color='#DEB887', alpha=0.3)  # BurlyWood color
plt.xlabel('Date', fontsize=14, fontweight='bold', color='#8B4513')
plt.ylabel('Price (USD)', fontsize=14, fontweight='bold', color='#8B4513')
plt.title('Bitcoin Price Over Time', fontsize=18, fontweight='bold', fontname='serif', color='#8B4513')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5, color='#D2B48C')  # Tan color
plt.xticks(fontsize=12, fontname='serif', color='#8B4513')
plt.yticks(fontsize=12, fontname='serif', color='#8B4513')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_color('#8B4513')
plt.gca().spines['bottom'].set_color('#8B4513')
plt.gca().tick_params(axis='x', colors='#8B4513')
plt.gca().tick_params(axis='y', colors='#8B4513')
# [END OF CUSTOMIZATION CODE]
plt.savefig('bitcoin_price_vintage.png', transparent=True)

plt.show()
```
